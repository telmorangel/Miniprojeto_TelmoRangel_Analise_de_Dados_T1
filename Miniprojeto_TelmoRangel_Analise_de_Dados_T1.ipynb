## 💡 Principais Insights e Conclusões
Com base nas etapas de limpeza e análise descritiva desenvolvidas neste projeto, observamos que:
1. **Auditoria e Limpeza de Dados:** Registros corrompidos ('#N/D', strings de espaços em branco) e uma coluna fantasma ('Unnamed') foram localizados e removidos, garantindo a integridade dos dados.
2. **Qualidade de Dados (Duplicatas):** A base original possuía 96.553 registros 100% duplicados. Sua remoção reduziu o dataset de 830.000 para 733.447 entradas, mitigando distorções analíticas e garantindo unicidade.

7. **Comportamento de Compra por Gênero:** Clientes do gênero feminino ('F') apresentaram um volume de compras ligeiramente superior ao masculino ('M'), sugerindo oportunidades para campanhas de marketing segmentadas.
8. **Padrões de Compra por Segmento:** O segmento de clientes 'B' demonstrou o maior volume de compras, indicando ser o principal target demográfico da empresa.
9. **Perfil de Compra por Estado Civil:** Clientes 'Separados' e 'Solteiros' lideraram em volume de compras, seguidos por 'Divorciados' e 'Casado ou união estável', com 'Viúvos' representando a menor parcela.
10. **Comportamento do Consumidor (Número de Filhos):** A análise da coluna 'CL_FHL' revelou que a maioria dos clientes (moda e mediana 0) não possui filhos ou tem um número baixo (média de 1.15, 75% dos clientes com até 2 filhos). Isso sugere que estratégias de marketing focadas em famílias grandes podem precisar de um nicho específico.
11. **Validação da Regra de Negócio (CO_ID):** A análise confirmou que a vasta maioria das compras (aproximadamente 98.8%) contém múltiplos itens, validando 'CO_ID' como identificador de cupom fiscal ou transação que agrega diversos produtos.

```python
# importação das bibliotecas necessárias ao projeto
import os
import csv
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
```

```python
# Realizando download da base via Kaggle API...")
path = kagglehub.dataset_download("namespaiva/base-varejo")
caminho_completo = os.path.join(path, "Base Varejo.csv")

dados_processados = []

# Leitura e extração estruturada e nativa com csv.DictReader
with open(caminho_completo, mode='r', encoding='utf-8') as arquivo:
    leitor_csv = csv.DictReader(arquivo, delimiter=';')
    leitor = list(leitor_csv)


df_varejo = pd.DataFrame(leitor)
```

# Transformação de Tipos


```python
# Lista das colunas que devem ser convertidas para numérico.
# 'CL_ID', 'CL_EC', 'CL_FHL' e 'PR_ID' são identificadores ou contagens que precisam de tratamento numérico.
colunas_numericas = ['CL_ID','CL_EC', 'CL_FHL','PR_ID']

# Itera sobre cada coluna na lista de colunas numéricas.
for coluna in colunas_numericas:
    # Converte a coluna atual para tipo numérico.
    # errors='coerce' é crucial: ele transformará quaisquer valores que não puderem ser convertidos
    # para numérico (ex: texto) em NaN (Not a Number), evitando erros e permitindo tratamento posterior.
    df_varejo[coluna] = pd.to_numeric(df_varejo[coluna], errors='coerce')

# Imprime uma linha separadora para melhorar a legibilidade da saída.
print("\n" + "="*50 + "\n")

# Exibe as informações atualizadas do DataFrame para verificar os novos tipos de dados.
# Espera-se que as colunas listadas em 'colunas_numericas' agora apareçam como int64 ou float64.
display(df_varejo.info())
```

```python
# 1. Converte a coluna de texto para o TIPO real 'datetime'
# O parâmetro dayfirst=True garante que o Pandas entenda que o primeiro número é o Dia (padrão português/brasileiro)
df_varejo["DATA"] = pd.to_datetime(df_varejo["DATA"], dayfirst=True, errors="coerce")

print("--- DataFrame Após Conversão para Datetime ---")
display(df_varejo.head())

# Ao fazer o .info(), a coluna DATA deve aparecer como 'datetime64[ns]' e NÃO como 'object'
display(df_varejo.info())
#print("\n" + "="*50 + "\n")
```

# Limpeza de Nulos e Duplicatas

```python

display(df_varejo.info())


# Remove coluna sem nome
print(df_varejo.columns.tolist())
df_varejo = df_varejo[['DATA','CO_ID','CL_ID','CL_GENERO', 'CL_EC',	'CL_FHL','CL_SEG','PR_ID','PR_CAT','PR_NOME']]
print("\n" + "="*50 + "\n")

# verificação de valores nulos nas colunas
print(df_varejo.isnull().sum())

# Exibe informações gerais sobre o DataFrame e suas dimensões.
print("--- Informações do DataFrame ---")
df_varejo.info()
print(f"\nDimensões do DataFrame: {df_varejo.shape[0]} linhas e {df_varejo.shape[1]} colunas")
print("\n" + "="*50 + "\n")

# Verifica e reporta linhas completamente duplicadas.
print("--- Verificação e Remoção de Duplicatas ---")
initial_duplicates = df_varejo.duplicated().sum()
print(f"Número de linhas 100% duplicadas encontradas: {initial_duplicates}")

# Remove linhas duplicadas e redefine o índice.
if initial_duplicates > 0:
    df_varejo = df_varejo.drop_duplicates().reset_index(drop=True)
    print(f"Linhas duplicadas removidas. Total de linhas restantes: {df_varejo.shape[0]}")
else:
    print("Nenhuma linha duplicada encontrada.")

print(f"Linhas 100% duplicadas após remoção: {df_varejo.duplicated().sum()}")
print("\n" + "="*50 + "\n")



```

```python
# Tradução dos códigos numéricos mascarados com base no dicionário oficial do projeto
dicionario_ec = {
    1: "Casado ou união estável",
    2: "Divorciado",
    3: "Separado",
    4: "Solteiro",
    5: "Viúvo",
}
df_varejo["CL_ESTADO_CIVIL"] = (
    df_varejo["CL_EC"].map(dicionario_ec).fillna("Não Informado")
)


# Validação final do processo de limpeza nesta etapa
print("\nMapeamento de nulos após aplicação das regras de tratamento:")
display(df_varejo.isnull().sum())
print("=" * 70)
df_varejo["CL_ESTADO_CIVIL"].sample(n=5)

```

```python
# Higienização das Colunas de Produto (PR_CAT e PR_NOME)


# Diagnóstico de categorias e nomes antes da higienização condicional
print("\n--- VALORES ÚNICOS EM 'PR_CAT' ANTES DA LIMPEZA ---")
display(df_varejo["PR_CAT"].value_counts(dropna=False))

print("\n--- VALORES ÚNICOS EM 'PR_NOME' ANTES DA LIMPEZA ---")
display(df_varejo["PR_NOME"].value_counts(dropna=False))


# --- TRATAMENTO CONDICIONAL DAS COLUNAS PR_CAT E PR_NOME ---

# Preenche os valores nulos estruturais nativos (NaN)
df_varejo["PR_CAT"] = df_varejo["PR_CAT"].fillna("Sem Categoria")
df_varejo["PR_NOME"] = df_varejo["PR_NOME"].fillna("Sem Nome")

# Corrige erros de texto '#N/D' e '# N/D' herdados de falhas em planilhas eletrônicas
# A lista ["#N/D", "# N/D"] garante a limpeza mesmo se houver variações de espaçamento
df_varejo["PR_CAT"] = df_varejo["PR_CAT"].replace(["#N/D", "# N/D"], "Sem Categoria")
df_varejo["PR_NOME"] = df_varejo["PR_NOME"].replace(["#N/D", "# N/D"], "Sem Nome")

# Trata strings compostas apenas por espaços vazios em branco
# Resumo da regra: "Encontre qualquer texto que, desde o seu início (^) até o seu fim ($), contenha # apenas espaços em branco (\s*) ou não tenha caractere nenhum."

df_varejo["PR_CAT"] = df_varejo["PR_CAT"].replace(r"^\s*$", "Sem Categoria", regex=True)
df_varejo["PR_NOME"] = df_varejo["PR_NOME"].replace(r"^\s*$", "Sem Nome", regex=True)

# Sertificar na troca correta do nome e categoria
print("\n--- AMOSTRA DE LINHAS 'Sem Nome' e 'Sem Categoria' ---")
display(df_varejo[df_varejo["PR_NOME"] == "Sem Nome"].head())

print("=" * 70)


display("Valores nulos e inconsistências como '#N/D' e strings vazias nas colunas 'PR_CAT' e 'PR_NOME' foram unificados sob rótulos como 'Sem Categoria' e 'Sem Nome', adiando a decisão de remoção definitiva dessas linhas que pouco contribuem.")

```

```python
# Verificação de Inconsistências entre PR_NOME e PR_ID

print("\n" + "=" * 70)
print("VERIFICAÇÃO DE INCONSISTÊNCIAS: PR_NOME vs PR_ID")
print("=" * 70)

# Agrupar por PR_NOME e coletar os IDs únicos para cada nome
inconsistencias_pr = df_varejo.groupby('PR_NOME')['PR_ID'].nunique()

# Filtrar apenas os produtos que possuem mais de um ID único
produtos_com_multiplos_ids = inconsistencias_pr[inconsistencias_pr > 1]

# Verificar se há inconsistências
if not produtos_com_multiplos_ids.empty:
    print("--- PRODUTOS COM O MESMO NOME MAS MÚLTIPLOS PR_ID's ---")
    print("\nForam encontradas inconsistências entre PR_NOME e PR_ID, onde o mesmo nome de produto está associado a diferentes códigos de produto. Isso pode indicar:")
    print("  - Variações do mesmo produto registradas com IDs diferentes.")
    print("  - Erros de digitação ou categorização na entrada dos dados.")
    print("  - Produtos semelhantes mas distintos que foram agrupados erroneamente pelo nome.")
    print("\nDetalhes das inconsistências (PR_NOME e contagem de PR_ID únicos):")
    display(produtos_com_multiplos_ids)

print("\n" + "=" * 70 + "\n")
```

```python
# Resolução de Inconsistências: Criação da Coluna PR_NOME_SEGMENTADO

# Nesta etapa, aplicamos uma função para garantir que produtos com o mesmo nome, mas IDs diferentes, sejam diferenciados.
def gerar_nome_segmentado(df):
    """
    Cria a coluna PR_NOME_SEGMENTADO.
    Se um PR_NOME possui mais de um PR_ID único, concatena o ID ao nome.
    """
    # Identifica produtos com múltiplos IDs
    contagem_ids = df.groupby('PR_NOME')['PR_ID'].nunique()
    nomes_inconsistentes = contagem_ids[contagem_ids > 1].index

    # Aplica a lógica de concatenação
    df['PR_NOME_SEGMENTADO'] = df.apply(
        lambda x: f"{x['PR_NOME']} (ID {x['PR_ID']})" if x['PR_NOME'] in nomes_inconsistentes else x['PR_NOME'],
        axis=1
    )
    return df

# Aplicando a função ao dataframe
df_varejo = gerar_nome_segmentado(df_varejo)

print("--- Coluna 'PR_NOME_SEGMENTADO' criada com sucesso ---")
display(df_varejo[['PR_ID', 'PR_NOME', 'PR_NOME_SEGMENTADO']].drop_duplicates().head(10))
```

# Estatística Descritiva

```python
# Estatística Descritiva
print("\n" + "=" * 70)
print("ANÁLISE OPERACIONAL POR AGRUPAMENTO (GROUPBY)")
print("=" * 70)

print("\n--- COMBINAÇÃO 1: VOLUME DE COMPRAS POR GÊNERO ---")
# Agrupa o DataFrame pela coluna 'CL_GENERO' e conta o número de ocorrências (compras) para cada gênero.
# reset_index(name='TOTAL_COMPRAS') transforma a Series resultante em um DataFrame com a coluna de contagem renomeada.
agrupamento_genero = (
    df_varejo.groupby("CL_GENERO").size().reset_index(name="TOTAL_COMPRAS")
)
# Exibe o resultado do agrupamento de forma formatada.
print(agrupamento_genero.to_string(index=False))

print("\n--- COMBINAÇÃO 2: VOLUME DE COMPRAS POR SEÇÃO/SEGMENTO ---")
# Similarmente, agrupa o DataFrame pela coluna 'CL_SEG' (segmento do cliente) e conta as compras.
agrupamento_segmento = (
    df_varejo.groupby("CL_SEG").size().reset_index(name="TOTAL_COMPRAS")
)
# Exibe o resultado.
print(agrupamento_segmento.to_string(index=False))

print("\n--- COMBINAÇÃO 3 (BÓNUS): VOLUME DE COMPRAS POR ESTADO CIVIL ---")
# Agrupa o DataFrame pela coluna 'CL_ESTADO_CIVIL' (já higienizada) e conta as compras.
agrupamento_ec = (
    df_varejo.groupby("CL_ESTADO_CIVIL").size().reset_index(name="TOTAL_COMPRAS")
)
# Exibe o resultado, ordenando pela contagem total de compras em ordem decrescente para destacar os maiores volumes.
print(
    agrupamento_ec.sort_values(by="TOTAL_COMPRAS", ascending=False).to_string(
        index=False
    )
)
```

```python
# Visualização Gráfica dos Agrupamentos
# Nesta seção, transformamos os dados agrupados em gráficos para facilitar a interpretação dos # #  # volumes de venda por categoria de cliente.

# Configuração do estilo visual
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1: Gênero
sns.barplot(data=agrupamento_genero, x='CL_GENERO', y='TOTAL_COMPRAS', ax=axes[0], palette='viridis', hue='CL_GENERO', legend=False)
axes[0].set_title('Compras por Gênero')
axes[0].set_xlabel('Gênero')
axes[0].set_ylabel('Total de Compras')

# Gráfico 2: Segmento
sns.barplot(data=agrupamento_segmento, x='CL_SEG', y='TOTAL_COMPRAS', ax=axes[1], palette='magma', hue='CL_SEG', legend=False)
axes[1].set_title('Compras por Segmento (Seção)')
axes[1].set_xlabel('Segmento')
axes[1].set_ylabel('')

# Gráfico 3: Estado Civil
agrupamento_ec_sorted = agrupamento_ec.sort_values(by='TOTAL_COMPRAS', ascending=False)
sns.barplot(data=agrupamento_ec_sorted, x='CL_ESTADO_CIVIL', y='TOTAL_COMPRAS', ax=axes[2], palette='rocket', hue='CL_ESTADO_CIVIL', legend=False)
axes[2].set_title('Compras por Estado Civil')
axes[2].set_xlabel('Estado Civil')
axes[2].set_ylabel('')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Estatística Descritiva da Coluna 'Número de Filhos'

print("\n" + "=" * 70)
print("ESTATÍSTICA DESCRITIVA - NÚMERO DE FILHOS (CL_FHL)")
print("=" * 70)

# Calcula as estatísticas descritivas da coluna 'CL_FHL' (número de filhos).
# O parâmetro percentiles=[0.25, 0.5, 0.75] adiciona os quartis 25%, 50% (mediana) e 75% à saída.
estatisticas = df_varejo["CL_FHL"].describe(percentiles=[0.25, 0.5, 0.75])

# Calcula a moda (o valor que mais aparece) da coluna 'CL_FHL'.
# .mode()[0] é usado para obter o primeiro valor caso haja múltiplas modas.
moda_filhos = df_varejo["CL_FHL"].mode()[0]

# Imprime as estatísticas descritivas de forma formatada.
print(f"Contagem Total:         {estatisticas['count']:.0f} clientes")
print(f"Média:                  {estatisticas['mean']:.2f}")
print(f"Mediana (Q2):           {estatisticas['50%']:.2f}")
print(f"Desvio Padrão:          {estatisticas['std']:.2f}")
print(f"Moda (Mais comum):      {moda_filhos:.0f}")
print(f"Mínimo registrado:      {estatisticas['min']:.0f}")
print(f"Máximo registrado:      {estatisticas['max']:.0f}")
print(f"1º Quartil (25%):       {estatisticas['25%']:.2f}")
print(f"3º Quartil (75%):       {estatisticas['75%']:.2f}")
```
